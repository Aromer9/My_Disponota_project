```python
import pandas as pd
import datetime
data = pd.read_csv('/Users/aromerq/Documents/Analitica y BI/Análisis/Capacidad Operacional/Instock/data.csv',  sep=';')
```

```python
#creo columnas necesarias
data['Instock'] = data['INSTOCK_TOTAL_SALA'] / data['TOTAL_SALA']
data['month'] = pd.to_datetime(data['Instock_Date']) 
data['month'] = data['month'].dt.month 

data['year'] = pd.to_datetime(data['Instock_Date']) 
data['year'] = data['year'].dt.year
```

```python
lider = data[data['Formato'].isin(['819  LIDER'])]
express = data[data['Formato'].isin(['826  EXPRESS'])]
sba = data[data['Formato'].isin(['815  BOD ACUENTA'])]
```

```python
df_lider = lider.groupby(['year','Formato']).mean()['Instock']
df_express = express.groupby(['year', 'Formato']).mean()['Instock']
df_sba = sba.groupby(['year', 'Formato']).mean()['Instock']
```
