```python
data['month'] = pd.to_datetime(data['fecha']) 
data['month'] = data['month'].dt.month 

data['year'] = pd.to_datetime(data['fecha']) 
data['year'] = data['year'].dt.year

data['quarter'] = pd.to_datetime(data['fecha'])
data['quarter'] = data['quarter'].dt.quarter
```

```python
#GRAFICO DE LA DISPONOTA TOTAL 
import matplotlib.pyplot as plt
import numpy as np

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(14, 8))

#FILTROS DE DATA
df_lider = data[data['cod_formato'].isin(['819'])]
df_express = data[data['cod_formato'].isin(['826'])]
df_express_400 = data[data['cod_formato'].isin(['200'])]
df_sba = data[data['cod_formato'].isin(['815'])]

#group BY
df_segmento1 = df_lider.groupby(['year', 'quarter']).mean()['disponota']
df_segmento2 = df_express.groupby(['year', 'quarter']).mean()['disponota']
df_segmento3 = df_express_400.groupby(['year', 'quarter']).mean()['disponota']
df_segmento4 = df_sba.groupby(['year', 'quarter']).mean()['disponota']


df_segmento1.plot(ax=ax1, kind='bar', color='b', ylabel='Disponota', xlabel='month/quarter', title='Media Disponota Lider')
df_segmento2.plot(ax=ax2, kind='bar', color='g', ylabel='Disponota', xlabel='month/quarter' , title='Media Disponota Express')
df_segmento3.plot(ax=ax3, kind='bar', color='g', ylabel='Disponota', xlabel='month/quarter' , title='Media Disponota Express 400')
df_segmento4.plot(ax=ax4, kind='bar', color='r', ylabel='Disponota', xlabel='month/quarter' , title='Media Disponota SBA')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.9)

plt.show()
```
