```python
#creo columnas necesarias
data['Fillrate'] = data['CAJAS_RECIBIDAS'] / data['CAJAS_ORDENADAS']
data['month']= pd.to_datetime(data['GREGORIAN_DATE']) 
data['month'] = data['month'].dt.month 
```

```python
data.head()
```

```python
#FILTRO POR FORMATO
df2 = data[data['FORMATO'].isin(['EXPRESS'])]
```

```python
df2 = df2[df2['month'].isin([1,2,3,4,5])]
```

```python
df2.head()
```

```python
df_segmento = df2.groupby(['AO', 'month', 'SEGMENTO']).mean()['Fillrate']
df_segmento
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize=(14, 8))

a2021 = df2[df2['AO'].isin([2021])]
a2021_ACP = a2021[a2021['SEGMENTO'].isin(['ACP'])]
a2021_GM = a2021[a2021['SEGMENTO'].isin(['GM'])]

a2022 = df2[df2['AO'].isin([2022])]
a2022_ACP = a2022[a2022['SEGMENTO'].isin(['ACP'])]
a2022_GM = a2022[a2022['SEGMENTO'].isin(['GM'])]

df_segmento1 = a2021_ACP.groupby(['AO', 'month', 'SEGMENTO']).mean()['Fillrate']
df_segmento2 = a2022_ACP.groupby(['AO', 'month', 'SEGMENTO']).mean()['Fillrate']
df_segmento3 = a2021_GM.groupby(['AO', 'month', 'SEGMENTO']).mean()['Fillrate']
df_segmento4 = a2022_GM.groupby(['AO', 'month', 'SEGMENTO']).mean()['Fillrate']


df_segmento1.plot(ax=ax1, kind='bar', color='b', ylabel='Fillrate', xlabel='month', title='Fillrate 2021')
df_segmento2.plot(ax=ax2, kind='bar', color='c', ylabel='Fillrate', xlabel='month' , title='Fillrate 2022')
df_segmento3.plot(ax=ax3, kind='bar', color='b', ylabel='Fillrate', xlabel='month' , title='Fillrate 2021')
df_segmento4.plot(ax=ax4, kind='bar', color='c', ylabel='Fillrate', xlabel='month' , title='Fillrate 2022')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.9)

plt.show()
```
